```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 

# Data
names = ['Siddarth', 'Tanmay', 'Abhas', 'Lakshya', 'Ambar']
marks = [78, 99, 88, 77, 90]

# Plotting a bar plot
plt.bar(names, marks)

# Labels and title
plt.xlabel("Names")
plt.ylabel("Marks")
plt.xticks(names)
plt.yticks(np.arange(0, 100, 5))
plt.title("Bar Plot of Marks")

# Display  the plot
plt.show()
```
